```python
dataset['seats'] = pd.to_numeric(dataset['seats'], errors='coerce')
dataset = dataset.dropna(subset=['seats'])
dataset['seats'] = dataset['seats'].astype(int)


dataset['max_power'] = pd.to_numeric(dataset['max_power'], errors='coerce')
dataset = dataset.dropna(subset=['max_power'])
dataset['max_power'] = dataset['max_power'].astype(float)

dataset = dataset.drop(['torque'],axis=1)
dataset.dtypes
```
